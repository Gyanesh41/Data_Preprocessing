# Write a Python program to draw a Scatter plot for  Country versus AvgTemperature using the dataframe created in (11).

```python
plot=sns.barplot(data=AvgTemp_Top_10,x="Country",y="AvgTemperature")
plt.xticks(rotation=90)
mean=list(AvgTemp_Top_10.groupby("Country").mean()["AvgTemperature"])
for i in range(len(AvgTemp_Top_10["Country"].unique())):
    plt.text(x=i,y=round(mean[i-1],2)/2,s=str(round(mean[i-1],2)),ha='center',rotation=90)
```

# Write a Python program to display a Pie Chart for the Country attribute where AvgTemperature is equal to  75  as shown below.

```python
#Getting list of Country whose  Average temperature = 75
Country_Avg_Temp_75=df.loc[(df["AvgTemperature"]==75),["Country","AvgTemperature"]]

#Getting How many times it of every country's average temperature = 75 
Country_Count={}
for i in Country_Avg_Temp_75["Country"].unique():
    Country_Count.update({i:Country_Avg_Temp_75.loc[Country_Avg_Temp_75["Country"]==i,"Country"].count()})
Country_Count=pd.DataFrame({"Country":Country_Count.keys(),
                           "Count":Country_Count.values()})

#Getting top 10 countrieswhose 75 temperature count is high
Country_Count_10=Country_Count.nlargest(10,"Count")

#Plot the pie
plt.pie(data=Country_Count_10,x=Country_Count_10["Count"],labels=Country_Count_10["Country"],textprops={'fontsize': 8},startangle=10,autopct='%1.1f%%')
plt.title("All Countries with no of times avg temp=75",fontdict={'size':15})
plt.figure(figsize=(50,50))
```

```python
sns.barplot(data=df,x='AvgTemperature',y='Month')
argdict={'charttype':'bar','data':df,'x':'AvgTemperature','y':'Month'}
def text(argdict):
    if argdict['charttype'] =='bar':
        mean=argdict['data'].groupby(argdict['x']).mean()[argdict['y']].reset_index();print(mean)
        for i in range(len(argdict['x'])):
            plt.text(x=argdict['x'][i],y=int((max(mean[argdict['y']])+min(mean[argdict['y']]))/2),s=round(mean[i],2),ha='centre')
    elif argdict['charttype'] in ['count','hist']:
        count=argdict['data'].groupby(argdict['x']).count()[argdict['y']].reset_index(inplace=True)
        for i in range(len(argdict['x'])):
            plt.text(x=argdict['x'][i],y=int((max(count[argdict['y']])+min(count[argdict['y']]))/2),s=round(count[i],2),ha='centre')
text(argdict)
```
