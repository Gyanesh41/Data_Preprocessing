```python
test=df.loc[pd.isna(df['Recovered'])]
train=df.loc[pd.isna(df['Recovered'])==False].dropna()
test.shape[0],train.shape[0]
```

## Most Confirmed Cases (Country Wise)

```python
df_c=df_l.groupby('Country_Region',as_index=False)['Confirmed'].mean().nlargest(10,columns='Confirmed')
df_c
```

```python
no1=lambda p: '{:.0f} Lakhs+ '.format(p *m1  / 100)
explode=[0.05,0,0,0,0,0,0,0,0,0]
```
